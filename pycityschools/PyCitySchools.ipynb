```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "C:/Users/ca25935/Desktop/UCD Data Analytics/Homework/Homework #4/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "C:/Users/ca25935/Desktop/UCD Data Analytics/Homework/Homework #4/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["Student ID"])
total_budget = school_data["budget"].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
overall_passing_rate = (avg_math_score + avg_reading_score) / 2

students_passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70,:]
students_passing_math = students_passing_math["Student ID"].count()
percent_passing_math = (students_passing_math / total_students) * 100

students_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70,:]
students_passing_reading = students_passing_reading["Student ID"].count()
percent_passing_reading = (students_passing_reading / total_students) * 100


#create a summary table for District Summary
summary_table = pd.DataFrame({"Total Schools": [total_schools],
                              "Total Students": [total_students],
                              "Total Budget": [total_budget],
                              "Average Math Score": [avg_math_score],
                              "Average Reading Score": [avg_reading_score],
                              "% Passing Math": [percent_passing_math],
                              "% Passing Reading": [percent_passing_reading],
                              "% Overall Passing Rate": [overall_passing_rate]})

summary_table["Total Budget"] = summary_table["Total Budget"].map("${:,.2f}".format)
summary_table["% Passing Math"] = summary_table["% Passing Math"].map("{:.0f}%".format)
summary_table["% Passing Reading"] = summary_table["% Passing Reading"].map("{:.0f}%".format)
summary_table["% Overall Passing Rate"] = summary_table["% Overall Passing Rate"].map("{:.0f}%".format)


# summary_table
```

```python
school_df = school_data[["school_name","type"]]
```

```python

#count of students
student_count_by_school = student_data[["school_name","Student ID"]]

# student_count_by_school
group_student_count_by_school = student_count_by_school.groupby(["school_name"])
group_student_count_by_school.count().head(15)

#convert to dataframe
df_student_counts = pd.DataFrame(group_student_count_by_school["Student ID"].count())


#merge to main school df

schools_df1 = pd.merge(school_df, df_student_counts, on="school_name", how="left" )
schools_df1 = schools_df1.rename(columns={"Student ID":"Total Students","type":"School Type"})

```

```python
#count of budget by school
df_budget_by_school = school_data[["school_name","budget"]]


#merge budget to master dataframe
schools_df2 = pd.merge(schools_df1, df_budget_by_school, on="school_name", how="left" )

#rename budget column
schools_df2 = schools_df2.rename(columns={"budget":"Total School Budget"})


#calculate budget per student
schools_df2["Per Student Budget"] = schools_df2["Total School Budget"] / schools_df2["Total Students"]

```

```python
#avg math score by school
df_mathscore_by_school = student_data[["school_name","math_score"]]

grouped_df_mathscore_by_school = df_mathscore_by_school.groupby(["school_name"])
grouped_df_mathscore_by_school.mean().head(15)

#convert to groupby object to dataframe
df_school_mathscore = pd.DataFrame(grouped_df_mathscore_by_school["math_score"].mean())

#rename math avg column
df_school_mathscore = df_school_mathscore.rename(columns={"math_score":"Average Math Score"})

#merge with master dataframe
schools_df3 = pd.merge(schools_df2, df_school_mathscore, on="school_name", how="left" )

```

```python
#avg reading score by school
df_readingscore_by_school = student_data[["school_name","reading_score"]]


grouped_df_readingscore_by_school = df_readingscore_by_school.groupby(["school_name"])
grouped_df_readingscore_by_school.mean().head(15)

#convert to groupby object to dataframe
df_school_readingscore = pd.DataFrame(grouped_df_readingscore_by_school["reading_score"].mean())

#rename reading avg column
df_school_readingscore = df_school_readingscore.rename(columns={"reading_score":"Average Reading Score"})


#merge with master dataframe
schools_df4 = pd.merge(schools_df3, df_school_readingscore, on="school_name", how="left" )
```

```python
#num of students passing math score by school
df_PercentPassMath_by_school = student_data.loc[student_data["math_score"]>=70,:]


#count of students
df_PercentPassMath_by_school = df_PercentPassMath_by_school[["Student ID","school_name"]]
df_PercentPassMath_by_school

# # student_count_by_school
group_studentpassmath_by_school = df_PercentPassMath_by_school.groupby(["school_name"])
group_studentpassmath_by_school.count().head(15)

#convert to dataframe
df_studentpassmath_counts = pd.DataFrame(group_studentpassmath_by_school["Student ID"].count())
df_studentpassmath_counts = df_studentpassmath_counts.rename(columns={"Student ID":"Students Passing Math"})
df_studentpassmath_counts

# #merge to main school df

schools_df5 = pd.merge(schools_df4, df_studentpassmath_counts, on="school_name", how="left" )



schools_df5['% Passing Math'] = (schools_df5["Students Passing Math"] / schools_df5["Total Students"]) * 100
del schools_df5['Students Passing Math']
```

```python
#num of students passing reading score by school
df_PercentPassReading_by_school = student_data.loc[student_data["reading_score"]>=70,:]


#count of students
df_PercentPassReading_by_school = df_PercentPassReading_by_school[["Student ID","school_name"]]


# # student_count_by_school
group_studentpassreading_by_school = df_PercentPassReading_by_school.groupby(["school_name"])
group_studentpassreading_by_school.count().head(15)

#convert to dataframe
df_studentpassreading_counts = pd.DataFrame(group_studentpassreading_by_school["Student ID"].count())
df_studentpassreading_counts = df_studentpassreading_counts.rename(columns={"Student ID":"Students Passing Reading"})


# #merge to main school df

schools_df6 = pd.merge(schools_df5, df_studentpassreading_counts, on="school_name", how="left" )



schools_df6['% Passing Reading'] = (schools_df6["Students Passing Reading"] / schools_df6["Total Students"]) * 100
del schools_df6['Students Passing Reading']
```

# School Summary

```python
#calculate % Overall Passing Rate


schools_df6['% Overall Passing Rate'] = (schools_df6["% Passing Math"] + schools_df6["% Passing Reading"]) / 2

schools_df6
```
